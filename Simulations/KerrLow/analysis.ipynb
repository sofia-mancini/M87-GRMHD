```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("metrics_values.csv")
```

```python
plt.figure(figsize=(6,4))

# b2
plt.plot
plt.hist(df['b2'].dropna(), bins=30, color='red', edgecolor='black')
plt.title(r'$\beta_2$')
plt.xlabel(r'$\beta_2$')
plt.ylabel('Count')

out_b2 = "/project/u2grc/Sofia/Simulations/KerrLow/b2.png"
plt.savefig(out_b2, dpi=300, bbox_inches='tight')


plt.show()

plt.figure(figsize=(12,10))

# b2/b1
plt.subplot(2, 2, 1)
plt.hist(df['b2/b1'].dropna(), bins=30, color='steelblue', edgecolor='black')
plt.title(r'$|\beta_2|\,/\,|\beta_1|$')
plt.xlabel(r'$|\beta_2|\,/\,|\beta_1|$')
plt.ylabel('Count')

# b2/betas
plt.subplot(2, 2, 2)
plt.hist(df['b2/sum_betas'].dropna(), bins=30, color='darkorange', edgecolor='black')
plt.title(r'$|\beta_2|\,/\,\sum |\beta_i|$')
plt.xlabel(r'$|\beta_2|\,/\,\sum |\beta_i|$')
plt.ylabel('Count')

# m_net
plt.subplot(2, 2, 3)
plt.hist(df['m_net'].dropna(), bins=30, color='seagreen', edgecolor='black')
plt.title('m_net (Net Linear Polarization)')
plt.xlabel('Fraction')
plt.ylabel('Count')

# m_avg
plt.subplot(2, 2, 4)
plt.hist(df['m_avg'].dropna(), bins=30, color='purple', edgecolor='black')
plt.title('m_avg (Average Linear Polarization)')
plt.xlabel('Fraction')
plt.ylabel('Count')

plt.tight_layout()

out_image = "/project/u2grc/Sofia/Simulations/KerrLow/histograms.png"
plt.savefig(out_image, dpi=300, bbox_inches='tight')

plt.show()
```
